```python
#Import necessary packages
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.datasets import load_digits 
import matplotlib.pyplot as plt
from sklearn import metrics
import seaborn as sn
import numpy as np
```

```python
#Load data to dataframe
dataset = load_digits()
print(dataset.data.shape)
print(dataset.target.shape)
```

```python
#Plotting the images in the digits dataset
plt.figure(figsize = (20,4))
for index , (image,label) in enumerate(zip(dataset.data[0:5], dataset.target[0:5])):
    plt.subplot(1,5, index+1)
    plt.imshow(np.reshape(image, (8,8)), cmap = plt.cm.gray)
    plt.title(label)
    
```

```python
#Split to train and test dataset

x_train, x_test, y_train, y_test = train_test_split(dataset.data,dataset.target, test_size=0.3,random_state=0 )
print(x_train.shape)
print(x_test.shape)
```

```python
#Instantiate the logistic regression model
logisticModel = LogisticRegression(max_iter=3000)

#Fit the logistic model with the train data
logisticModel.fit(x_train,y_train)
```

```python
predicted_values= logisticModel.predict(x_test)
score = logisticModel.score(x_test,y_test)
print(score)
```

```python
#Confusion matrix 
cm = metrics.confusion_matrix(y_test,predicted_values)
plt.figure(figsize=(10,10))
sn.heatmap(cm, annot=True, fmt =".3f", linewidths=.5, square =True,cmap = 'Blues_r')
plt.ylabel("Actual Value")
plt.xlabel("Predicted Values")
all_sample_title='Accuracy Score: {0}'.format(score)
plt.title(all_sample_title)
```

```python
# Instatite Gaussian Bayes Model
gaussian_model= GaussianNB()
gaussian_model.fit(x_train,y_train)
#Evaluate the performance of the Gaussian Naive Bayes
gauss_predicted_values = gaussian_model.predict(x_test)
gaus_accuracy_score = gaussian_model.score(x_test,y_test)
print(gaus_accuracy_score)
```

```python
#Bernouli Bayes
bernouli_model= BernoulliNB()
bernouli_model.fit(x_train, y_train)

#Evaluate the performance
bernouli_predicted_values = bernouli_model.predict(x_test)
beurnouli_acc = bernouli_model.score(x_test,y_test)
print(beurnouli_acc)
```

```python
#Multinomial Bayes
multinomial_model= MultinomialNB()
multinomial_model.fit(x_train, y_train)

#Evaluate the performance
multinomial_predicted_values = multinomial_model.predict(x_test)
multinomial_acc = multinomial_model.score(x_test,y_test)
print(multinomial_acc)
```
